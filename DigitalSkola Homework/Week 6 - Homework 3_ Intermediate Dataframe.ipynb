```python
import numpy as py
import pandas as pd
```

```python
data1 = pd.read_csv('https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data.csv')
```

```python
data1.shape
```

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data_2.csv')
```

```python
# Full outer dataset 1 dan dataset 2
data_merge = data1.merge(data2,how='outer',on = 'employee_id')
data_merge.head()
```

```python
# Display dataframe yang tidak match (awards_won_y == Nan)
not_match = data_merge[data_merge['awards_won_y'].isnull()]
not_match
```
