Diberikan sebuah list sebagai input dengan 10 item, buatlah logic untuk print 
hasil penjumlahan dimana item dari list bukan merupakan angka 5 dan 6

Cara untuk mendapat kan list input:
sertakan code
import random
input_list= [random.randrange(1, 50, 1) for i in range(10)]

Contoh:
input_list = [1,2,2] -> 5

input_list = [1,2,5,6] -> 3


```python
import random
```

```python
input_list= [random.randrange(1, 50, 1) for i in range(10)]
```

```python
input_list
```

```python
input_list.sort()
```

```python
print(input_list)
```

```python
sum(input_list[1:]) 
# karna tidak ada angka dibawah angka 6, maka list disortir dan penjumlahan menggunakan sum dengan range dimulai dari index 1
#jika ada angka dibawah angka 6, angka 6 harus dihapus terlebih dahulu dengan cara input_list.remove(6) kemudian sum seperti biasa
```
