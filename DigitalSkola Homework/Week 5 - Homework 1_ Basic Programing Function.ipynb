```python
def phalindrome(input_list):
  temp = input_list
  reverse_num = 0
  while(input_list>0):
    digit=input_list%10
    reverse_num=reverse_num*10+digit
    input_list=input_list//10

  if (temp==reverse_num):
    return 'bilangan input merupakan bilangan palindrom'
  else:
    return 'bilangan input bukan bilangan palindrom'
```
